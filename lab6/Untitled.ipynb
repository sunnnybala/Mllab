```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os
```

```python
data = {
    'salary': [1.7, 2.4, 2.3, 3.1, 3.7, 4.2, 4.4, 6.1, 5.4, 5.7, 6.4, 6.2],
    'experience': [1.2, 1.5, 1.9, 2.2, 2.4, 2.5, 2.8, 3.1, 3.3, 3.7, 4.2, 4.4]
}

df = pd.DataFrame(data)
```

```python
def forwardprop(data, param):
    y = []
    temp =0
    X = data.to_numpy()
    b = param[-1]
    W = param[:-1]
    Y = X@ (W.T) + b
    return Y
```

```python
def backprop(data, param, lr= 0.01, sub = 'a'):
    if sub == 'a':
        param[:-1] = param[:-1]+ lr
    elif sub == 'b':
        param[-1] = param[-1]+ lr
    return param
```

```python
def model(df, epochs, learn =0.01,suba='a', shcost =False):
    params =[]
    errors = []
    for i in df.iloc[:,:-1].columns:
        params.append(np.random.rand())
    params.append(np.random.rand())
    params = np.array(params)
    params[:-1] = 0
    params[-1] = 1.1
    for i in range(epochs):
        Y_pred = forwardprop(df.iloc[:,:-1], params)
        y_act = df.iloc[:,-1].to_numpy()
        err = np.sum((y_act - Y_pred)**2) / len(y_act)
        errors.append(err)
        params = backprop(df, params, lr =learn, sub=suba)
        print(params)
    return params, errors
```

```python
params, errors = model(df, 150, learn =0.01,suba = 'a', shcost =True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(errors, color='green', marker='o', linestyle='-')
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error')
plt.title('Error vs. Iterations')
plt.show()
```
