```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = [0,20,40,60,80,100]
y = [2*i for i in x]
ax.set_yticks([0, 50, 100,150,200])
ax.plot(x, y)

ax.set_xlabel('X Axis Label')  
ax.set_ylabel('Y Axis Label')  
ax.set_title('Plot Title')     

plt.show()
```

```python
adv = pd.read_csv('company_sales_data.csv')
display(adv)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
prof = adv['total_profit']
mon = adv['month_number']
ax.plot(mon, prof)
ax.set_title('Q_3 PLOT')
ax.set_xlabel('month_number')
ax.set_ylabel('total_profit')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
prof = adv['total_units']
mon = adv['month_number']
ax.plot(mon, prof, linestyle='--', color='red', marker='o', markerfacecolor='red', linewidth=3, label='sold Units')
ax.set_title('Q_4 PLOT')
ax.set_xlabel('month_number')
ax.set_ylabel('sold_units')
ax.legend(loc='lower right')
plt.show()
```

```python

plt.figure(figsize=(10, 6))

fig, ax = plt.subplots()
products = adv.columns[1:7]
for product in products:
    ax.plot(adv['month_number'], adv[product], marker='o', label=product)

plt.title('Units Sold per Month for Each Product')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(title='Products', loc='upper left')
plt.xticks(rotation=45)  

plt.show()
```
